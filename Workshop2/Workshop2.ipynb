# <b> Lists </b>

A list is an **ordered**, **mutable** and **dynamic** collection of items enclosed within square brackets `[]`.

Understanding Python lists can be made easier by comparing them to something familiar, like a grocery list.


### **1. Ordered Collection of Items**
- **Grocery List:** When you write a grocery list items are in a specific order.

- **Python List:** Similarly, a Python list is an ordered collection where each element has a specific position (index).


````python
# Python list of grocery items
groceries = ['milk', 'bread', 'eggs', 'apples']
````

### **2. Mutable and Dynamic**

- **Grocery List:** You can add items to your grocery list, cross out items you no longer need, or rearrange items.

- **Python List:**
  - Python lists are ***mutable meaning you can modify them after creation during runtime***
  - They are considered ***dynamic given that lists don't have a fixed length.***  

````python
# Adding an item to the list
groceries.append('bananas')
print(groceries)

# Removing an item from the list
groceries.remove('bread')
print(groceries)

# Changing an item in the list
groceries[2] = 'free-range eggs' # MUTABILITY
print(groceries)
````

### **3. Different types of Items**
- **Grocery List:** Your grocery list might include a variety of items: produce, dairy, bakery goods, etc.

- **Python List:** Python lists can contain elements of different data types, including numbers, strings, and even other lists.

````python
# List with different types of items
mixed_groceries = ['milk', 12, True, ['apples', 'oranges']]
print(mixed_groceries)
````

### **4. Indexed Access**
- **Grocery List:** You can refer to items by their position on the list, like the first item or the last item.

- **Python List:** Each item in a list has an index starting from 0, allowing you to access items directly.

# **Sets**

Sets are unordered, unindexed collections with no duplicate elements.

- ### **Key Characteristics of Sets:**
  - Unordered: Elements do not have a defined order.
  - Unindexed: You cannot access items using indices.
  - No Duplicates: Each element is unique.
  - Mutable: You can add or remove elements.
  - Elements Must Be Immutable: Elements themselves must be of immutable data types.

Sets are defined by placing elements inside curly braces {} or by using the set() function.

````python
# to define an empty set
empty_set = set()
print(empty_set)

numbers = {1,2,3,4}
print(numbers)

# sets are also heterogenous
mixed_set = {1, "Hello", 3.14, True}
print(mixed_set)


````

### **Accessing Set Elements:**
Since sets are unordered and unindexed, you cannot access elements via indices. However, you can loop through the set.

````python
fruits = {'apple', 'banana', 'cherry'}

for fruit in fruits:
    print(fruit)
````

### **Checking for Membership:**
You can check if an item exists `in` a set using in and `not in`.



````python
print('apple' in fruits)     # Output: True
print('orange' not in fruits)  # Output: True
````

### **Adding and removing Elements:**

Sets in Python are mutable, allowing you to add or remove elements after their creation. However, one of the key features of a set is that it only contains unique elements. This means:

- No Duplicates Allowed: Adding an element that already exists in the set will have no effect.
- Automatic Deduplication: If you add multiple elements at once, any duplicates are automatically removed.





````python
# Initialize a set
my_set = {1, 2, 3}

# Add an element
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}

# Attempt to add a duplicate element
my_set.add(2)
print(my_set)  # Output: {1, 2, 3, 4}
````

````python
my_set.remove(2)
print(my_set)
````

# **Object Oriented Programming**


image.png




**Attributes:** are variables inside a class that describe an **Object**.
    -  also referred to as Properties.

**Methods:** are functions inside a class that describe an **Object**.

Classes act as a blueprint composed of **Attributes** and **Methods** that abstractly describe **objects**.
- Imagine Car Manufacturing Company Designing a Blueprint(Class) for their Vehicle.

**Objects** are multiple specific instances defined by the Class blueprint.
- Imagine Cars that have been designed following this Blueprint with specific features that respect the desires of the custommers.

An **instance** is a specific realization of a class; the terms "object" and "instance" are often used interchangeably.

OOP is one of many programming paradigms that was first introduced in 1967 by the language "Simula".

Most modern day programming languages including Python are multi-paradigm using aspects of "Functional", "Procedural" and "Object Oriented" Paradigms.


Here we have an example of the class `Employee`.


````python
class Employee:

    def __init__(self, first, last, pay):
    ################ Properties ################
        self.first = first # Access level : public
        self._last = last # Access level : protected
        self.__pay = pay # Access level : private

    ################ Methods ################
    def fullname(self):
        return self.first + " " + self._last

    def email(self):
        return self.first + "." + self._last + "@company.com"

    # Setter
    def set_pay(self, val):
        # validation
        if val < 0:
            raise TypeError("Pay can't be Negative") # tells Python to throw an 'Type' error with that message.
        self.__pay = val

    #getters
    def get_pay(self):
        return f'${self.__pay}'

foo  = Employee("Thevin","Silva",69)
foo.set_pay(420)
foo.get_pay()
````

- `__init__` : is refered to as the **Constructor** method.
    - It takes in parameters at **instantiation** which are assigned to the Attribute.
    - Constructors don't return a value.
    - It runs at **instantiation**.

- `self` : gives you access to **Instance variables**.
    - **Instance variables** are defined at instantiation and can be different for each object.

````python
bar  = Employee("Lumi","Long",420)
print(bar.first)
print(foo.first)
````

## **Inheritance:**

# Magic Methods (Dunder Methods) Overview

Magic methods, also known as dunder methods (short for "double underscore"), are special methods in Python that allow objects to interact with built-in functions and operators. Below is a table summarizing some common magic methods, organized by category, along with brief examples.

| **Category**                         | **Magic Method**        | **Description**                                             | **Example Usage**                           |
|--------------------------------------|-------------------------|-------------------------------------------------------------|---------------------------------------------|
| **Object Creation and Representation** | `__init__(self, ...)`    | Initializes a new instance of a class.                       | `def __init__(self, name): self.name = name` |
|                                      | `__str__(self)`         | Returns a readable string representation of the object.      | `def __str__(self): return self.name`       |
|                                      | `__repr__(self)`        | Returns an unambiguous string representation of the object.  | `def __repr__(self): return f"Person('{self.name}')"` |
| **Arithmetic Operators**             | `__add__(self, other)`  | Defines behavior for the addition operator `+`.              | `def __add__(self, other): return self.value + other.value` |
| **Comparison Operators**             | `__eq__(self, other)`   | Defines behavior for the equality operator `==`.             | `def __eq__(self, other): return self.id == other.id` |
| **Container and Sequence Methods**   | `__len__(self)`         | Defines behavior for the built-in `len()` function.          | `def __len__(self): return len(self.items)` |
| **Iterator Methods**                 | `__iter__(self)`        | Returns an iterator object.                                  | `def __iter__(self): return iter(self.items)` |
|                                      | `__next__(self)`        | Returns the next item from an iterator.                      | `def __next__(self): return next(self.iter)` |


````python
class Developer(Employee):
    def __init__(self,first,last, pay, language, project):
        super().__init__(first,last,pay) # calling the previous constructor with the parameters passed into the new constructor
        self.language = language
        self.__project = project # Access level : Private

    # Function Overwriting
    def email(self):
        return self.first + "." + self.last + "@company.dev.com"

    # property decorator makes the method behave
    @property
    def project(self):
        return self.__project

    @project.setter
    def project(self, project):
        self.__project = project

    # Controls
    def __str__(self):
        # NOTE : get_pay() instead of self.__pay as this is private and inherited from the super class.
        return f"""
            name: {self.fullname()}
            salary: {self.get_pay()}
            language: {self.language}
            project: {self.project}
        """

foo = Developer("Thevin", "Silva", "420","Rust","Workshop2")
foo.fullname() # using methods from parent class

foo.project = "something"
foo.project

print(foo)
````

- `super()` : refers to the parent class.
    - offers access to methods defined in the Parent class for example the Constructor (`__init__`)
    - Used to call the parent's constructor with the parameters passed into the new constructor
    - Attributes from parent class are still present in the child class

# Extra:

- `@property` : decorator used to implement a Getter Method
    - Transforms the method `project()` into a getter for a read-only attribute called project.
    - Allows you to access `self.__project` using `[name of the object].project` instead of `[name of the object].project()`.

- `@project.setter` : defines the setter method for the `project` property.
    - Allows you to set the value of `self.__project` using `object.project = value`.

## **Polymorphism**

- **Generic Description**: Polymorphism is the ability of different objects to be accessed through the same interface, allowing the same operation to behave differently on different classes.

- #### <u>Interfaces:</u>
    - Define a contract of methods and attributes that classes must implement.
        - In languages like C# and Java, this is achieved using **interfaces**.
        - In Python, **abstract base classes (ABCs)** fulfill the same role.

- #### <u>Duck Typing:</u>
    








image.png

- Relies on an object's methods and properties rather than its explicit type.
- Based on the concept: "If it looks like a duck and quacks like a duck, it's a duck."
- Allows for flexible code where objects of different types can be used interchangeably if they implement the required methods.
- Common in dynamically typed languages like Python.


````python

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        print("Bark")

class Cat(Animal):
    def speak(self):
        print("Meow")

````
